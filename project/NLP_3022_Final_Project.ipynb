## What is RAG? ü§î
We've all used ChatGPT and experienced moments when it starts to generate content that is often incorrect or unrelated to our query. Do you know why this happens? These Large Language Models (LLMs) are not magical entities; they are simply models trained on a vast amount of text. üìö You could even consider a significant portion of the internet. However, this is not all the data available in the world, because data is not a static concept. You yourself generate some data every day through your use of the Internet, Social Media, and so on. üåêüíªüì±

So, no matter how much data you use to train your LLM, you always end up encountering new data. This is one of the reasons behind the famous ChatGPT response that tells you it only knows things up to a certain date. üìÖ Also, these models tend to hallucinate too. It means they provide incorrect answers but in a very convincing manner. üé≠

On the other hand, we have retrieval techniques. Don't worry if it sounds complicated (it actually isn't easy, you may need to take a course to familiarize yourself with these concepts üòÖ, but that's not necessary for this project), but you use it on a daily basis. You can think of Search Engines (like Google, for example) as a complex form of information retrieval. üîç

So, one day, people came up with this idea that it would be cool if ChatGPT could search Google for us, read the articles for us, summarize what it read, and tell us that. üìñ So, this is not exactly what RAG is, but it's something similar. We have a corpus (a large amount of data) and a query (what a user typed as input). Now, we search through this corpus using techniques related to vectors and vector databases, and find the most similar items in our corpus to the query. Then, we pass these items to an LLM and ask for a structured, well-formatted, user-friendly output. üìàüìä

## Create a Retriever üïµÔ∏è‚Äç‚ôÇÔ∏è
Now, we‚Äôre going to create a simple retriever. The role of the retriever is to compare the user‚Äôs query with a large corpus of text and find those that are most similar in context. (You know what context is by now, don‚Äôt you? üòä If you‚Äôve forgotten, refer back to your initial lectures). For now, let‚Äôs say we want to find similar text based on simple similarity metrics. The code is straightforward, and I have faith in you, chief! Dive into the code. üë®‚Äçüíª

```python
# in practive `query` and `document` are vectors and we compute their inner product 
#   and top-k are the ones with maximum inner products

def jaccard_similarity(query, document):
    query = query.lower().split(" ")
    document = document.lower().split(" ")
    intersection = set(query).intersection(set(document)) # set -> remove frequent words
    union = set(query).union(set(document))
    return len(intersection)/len(union)
```

Hey, you may want to look at wikipedia page for [Jaccard Similarity](https://en.wikipedia.org/wiki/Jaccard_index).

```python
def return_response(query, corpus):
    similarities = []
    for doc in corpus:
        # similarity = jaccard_similarity(user_input, doc)
        similarity = jaccard_similarity(query, doc)

        similarities.append(similarity)
    return corpus_of_documents[similarities.index(max(similarities))] # argmax -> similarities.index(max(similarities))
```

## üõ†Ô∏è Create a Retriever
To create your retriever, you need to use an encoder model. Something like BERT? Nah, BERT is so yesterday. Find something new and shiny! ‚ú® The basic idea is to encode every document (sentence) in your corpus into a vector space using the same encoder. Then, encode the user query into that same space. With some similarity metrics like dot product, you can find the most similar document to the user‚Äôs input and retrieve it. üéØ You can train your own encoder if you have enough data and resources, üí™ or you can use one of those [ready-made on Hugging Face](https://huggingface.co/models?pipeline_tag=sentence-similarity&sort=trending), like these ones.

```python
!pip install -U sentence_transformers --quiet
```

```python
### claude

from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode the corpus
corpus_embeddings = model.encode(corpus, show_progress_bar=True)

def retrieve(query, top_k=1):
    query_embedding = model.encode(query)
    similarities = np.dot(corpus_embeddings, query_embedding)
    top_indices = np.argsort(similarities)[-top_k:][::-1]
    return [corpus[i] for i in top_indices]


# ### GPT 
# from sentence_transformers import SentenceTransformer, util

# # Load the model
# retriever_model = SentenceTransformer('all-MiniLM-L6-v2')

# # Encode the corpus
# encoded_corpus = retriever_model.encode(corpus_of_documents, convert_to_tensor=True)

# # Define a retrieval function
# def retrieve(query, encoded_corpus, corpus_of_documents, retriever_model):
#     query_embedding = retriever_model.encode(query, convert_to_tensor=True)
#     scores = util.pytorch_cos_sim(query_embedding, encoded_corpus)[0]
#     best_match_idx = scores.argmax()
#     return corpus_of_documents[best_match_idx]

# # Test the retriever
# query = "I want to watch an action-packed movie."
# retrieved_document = retrieve(query, encoded_corpus, corpus_of_documents, retriever_model)
# print(f"Retrieved document: {retrieved_document}")


# ### cluade Medical Advisor
# from sentence_transformers import SentenceTransformer
# import numpy as np

# model = SentenceTransformer('pritamdeka/S-PubMedBert-MS-MARCO')

# # Encode the corpus
# corpus_embeddings = model.encode(medical_corpus, show_progress_bar=True)

# def retrieve(query, top_k=1):
#     query_embedding = model.encode(query)
#     similarities = np.dot(corpus_embeddings, query_embedding)
#     top_indices = np.argsort(similarities)[-top_k:][::-1]
#     return [medical_corpus[i] for i in top_indices]

# # Test the retriever
# test_query = "What are the symptoms of diabetes?"
# print("Test query:", test_query)
# print("Retrieved medical info:", retrieve(test_query)[0])
```

```python
### claude

judge_model = pipeline("text-classification", model="facebook/bart-large-mnli")

results = []
for query in queries:
    retrieved_doc = retrieve(query)[0]
    response1 = generator1(f"Query: {query}\nReview: {retrieved_doc}\nRecommendation:")
    response2 = generator2(f"Query: {query}\nReview: {retrieved_doc}\nRecommendation:")
    response3 = generator3(f"Query: {query}\nReview: {retrieved_doc}\nRecommendation:")
    
    scores = []
    for response in [response1, response2, response3]:
        score = judge_model(f"Query: {query}\nResponse: {response}", candidate_labels=["relevant", "irrelevant"])
        scores.append(score[0]['score'] if score[0]['label'] == "relevant" else 0)
    
    best_model = ["Model 1", "Model 2", "Model 3"][np.argmax(scores)]
    results.append(best_model)

model_scores = {
    "Model 1": results.count("Model 1"),
    "Model 2": results.count("Model 2"),
    "Model 3": results.count("Model 3")
}

print("Final Scores:")
for model, score in model_scores.items():
    print(f"{model}: {score}")


# ### GPT 
# from transformers import GPT2ForSequenceClassification, GPT2TokenizerFast

# # Load judge model and tokenizer
# judge_model_name = "gpt2-medium"
# judge_tokenizer = GPT2TokenizerFast.from_pretrained(judge_model_name)
# judge_model = GPT2ForSequenceClassification.from_pretrained(judge_model_name)

# # Function to score responses
# def score_responses(responses):
#     scores = []
#     for response in responses:
#         inputs = judge_tokenizer(response, return_tensors="pt")
#         outputs = judge_model(**inputs)
#         score = outputs.logits.softmax(dim=-1)[0][1].item()
#         scores.append(score)
#     return scores

# # Evaluate responses
# all_scores = []
# for query in queries:
#     retrieved_document = retrieve(query, encoded_corpus, corpus_of_documents, retriever_model)
#     responses = []
#     for generator in generators:
#         prompt = f"You are a bot that makes recommendations for movies. Try to be a helpful recommender system.\nThis is the recommended movie: {retrieved_document}\nThe user input is: {query}\nCompile a recommendation to the user based on the recommended movie and the user input."
#         output = generator(prompt, max_length=100)
#         responses.append(output[0]['generated_text'])
#     scores = score_responses(responses)
#     all_scores.append(scores)

# # Calculate the final scores
# model_scores = [0, 0, 0]
# for scores in all_scores:
#     best_model_idx = scores.index(max(scores))
#     model_scores[best_model_idx] += 1

# for i, model_name in enumerate(generator_models):
#     print(f"Model: {model_name}, Score: {model_scores[i]}")

### cluade Medical Advisor
# from transformers import pipeline

# judge_model = pipeline("text-classification", model="facebook/bart-large-mnli")

# results = []
# for query in queries:
#     retrieved_info = retrieve(query)[0]
#     response1 = generator1(f"Medical question: {query}\nRelevant information: {retrieved_info}\nMedical advice:")
#     response2 = generator2(f"Medical question: {query}\nRelevant information: {retrieved_info}\nMedical advice:")
#     response3 = generator3(f"Medical question: {query}\nRelevant information: {retrieved_info}\nMedical advice:")
    
#     scores = []
#     for response in [response1, response2, response3]:
#         score = judge_model(f"Question: {query}\nResponse: {response}", candidate_labels=["relevant", "irrelevant"])
#         scores.append(score[0]['score'] if score[0]['label'] == "relevant" else 0)
    
#     best_model = ["Model 1 (BioGPT)", "Model 2 (FLAN-T5)", "Model 3 (GPT-Neo-125M)"][np.argmax(scores)]
#     results.append(best_model)

# model_scores = {
#     "Model 1 (BioGPT)": results.count("Model 1 (BioGPT)"),
#     "Model 2 (FLAN-T5)": results.count("Model 2 (FLAN-T5)"),
#     "Model 3 (GPT-Neo-125M)": results.count("Model 3 (GPT-Neo-125M)")
# }

# print("\nFinal Scores:")
# for model, score in model_scores.items():
#     print(f"{model}: {score}")
```
